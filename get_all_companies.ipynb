### Step 1 - find level 1 categories

```python
#url = 'https://www.trustpilot.com/categories'
#r=requests.get(url)
#soup = BeautifulSoup(r.text, 'lxml')
#hrefs = []
#for a in soup.find_all('a', href=True):
#    hrefs.append(a['href'])
lev_1_cats = ['/categories/animals_pets',
             '/categories/beauty_wellbeing',
             '/categories/business_services',
             '/categories/construction_manufactoring',
             '/categories/education_training',
             '/categories/electronics_technology',
             '/categories/events_entertainment',
             '/categories/food_beverages_tobacco',
             '/categories/health_medical',
             '/categories/hobbies_crafts',
             '/categories/home_garden',
             '/categories/home_services',
             '/categories/legal_services_government',
             '/categories/media_publishing',
             '/categories/money_insurance',
             '/categories/public_local_services',
             '/categories/restaurants_bars',
             '/categories/shopping_fashion',
             '/categories/sports',
             '/categories/travel_vacation',
             '/categories/utilities',
             '/categories/vehicles_transportation']
```

```python
full_list = []

for i in lev_1_cats:
    print(i)
    lev_2_list = []
    # Then loop through pages
    pg_count = 0
    page_test = 1
    while page_test == 1:
        pg_count += 1
        url = 'https://www.trustpilot.com' + i + '?page=' + str(pg_count)
        r=requests.get(url)
        soup = BeautifulSoup(r.text, 'lxml')
        review_href = []
        for a in soup.find_all('a', href=True):
            if '/review/' == a['href'][:8]:
                review_href.append(a['href'])        
        if len(review_href) == 0:
            page_test = 0
        else:
            lev_2_list.append(review_href)
            
        
    full_list.append([item for sublist in lev_2_list for item in sublist])
```
